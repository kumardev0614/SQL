<b>Q0.</b> How many customers are male and how many are female?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    customers.gender,
    COUNT(customers.gender) AS Count  
FROM customer_dim AS customers

GROUP BY customers.gender


---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q5.</b> For each Pin code, print the count of orders, sum of total amount paid, average amount paid, maximum amount paid, minimum amount paid for the transactions which were paid by 'cash'. Take only 'buy' order types

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    orders.delivery_pincode AS pincode,
    COUNT(orders.order_id) AS count_of_order,
    SUM(orders.total_amount_paid) AS Total_ammount_paid,
    AVG(orders.total_amount_paid) AS avg_amount_paid,
    MAX(orders.total_amount_paid) AS max_amount_paid,
    MIN(orders.total_amount_paid) AS min_amount_paid
FROM order_dim AS orders

WHERE order_type IS 'buy' AND payment_type IS 'cash'

GROUP BY orders.delivery_pincode

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q6.</b> For each delivery_person_id, print the count of orders and total amount paid for product_id = 12350 or 12348 and total units > 8. Sort the output by total amount paid in descending order. Take only 'buy' order types

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    orders.delivery_person_id AS delivery_person,
    COUNT(orders.order_id) AS count_of_order,
    SUM(orders.total_amount_paid) AS Total_ammount_paid
FROM order_dim AS orders

WHERE (orders.product_id = 12350 OR orders.product_id = 12348) AND (orders.tot_units > 8) AND (orders.order_type = 'buy')

GROUP BY orders.delivery_person_id

ORDER BY Total_ammount_paid DESC

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q7.</b> Print the Full names (first name plus last name) for customers that have email on "gmail.com"?

<b>Q8.</b> How many orders had #units between 1-3, 4-6 and 7+? Take only 'buy' order types

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    table1.unit_bucket,
    COUNT(table1.unit_bucket) AS order_counts

FROM
    (
    SELECT
        CASE WHEN orders.tot_units IS NULL THEN 'Missing'
            WHEN orders.tot_units BETWEEN 1 AND 3 THEN "1-3"
            WHEN orders.tot_units BETWEEN 4 AND 6 THEN "4-6"
        ELSE '7+' END AS unit_bucket
    
    FROM order_dim AS orders
 
    WHERE orders.order_type = "buy"
    )  AS table1

GROUP BY table1.unit_bucket

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q9.</b> Which pincode has average amount paid more than 150,000? Take only 'buy' order types

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    orders.delivery_pincode AS pincode,
    AVG(orders.total_amount_paid) AS avrg_amount_paid

FROM order_dim AS orders

WHERE orders.order_type = "buy"

GROUP BY orders.delivery_pincode

HAVING avrg_amount_paid > 150000

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q11.</b> How many total orders were there in each month and how many of them were returned? Add a column for return rate too.<br>
return rate = (100.0 * total return orders) / total buy orders<br>
Hint: You will need to combine SUM() with CASE WHEN

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------
SELECT 
    month_groupby.Month,
    month_groupby.total_buy,
    month_groupby.total_returns,
    ((100.0 * month_groupby.total_returns)/month_groupby.total_buy) || "%" AS return_rate
FROM
    (
    SELECT
        monthly_orders.Month,
        SUM(CASE WHEN monthly_orders.order_type = 'buy' then 1 Else 0 End) AS total_buy,
        SUM(CASE WHEN monthly_orders.order_type = 'return' then 1 Else 0 End) AS total_returns
    FROM 
        (
        SELECT
            orders.order_id,
            orders.order_type,
            SUBSTR(orders.order_date, 4,2) AS Month
        FROM order_dim AS orders
        ) AS monthly_orders
    
    GROUP BY monthly_orders.Month
    ) AS month_groupby
---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    SUBSTR(orders.order_date, 4,2) AS Month,
    SUM(CASE WHEN orders.order_type = 'buy' then 1 Else 0 End) AS total_buy,
    SUM(CASE WHEN orders.order_type = 'return' then 1 Else 0 End) AS total_returns,
    (100.0 * SUM(CASE WHEN orders.order_type = 'return' then 1 Else 0 End)) / SUM(CASE WHEN orders.order_type = 'buy' then 1 Else 0 End) AS return_rate

FROM order_dim AS orders

GROUP BY SUBSTR(orders.order_date, 4,2)

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<h2> Questions on SQL Joins</h2>

<b>Q12.</b> How many units have been sold by each brand? Also get total returned units for each brand.

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    products.brand,
    SUM(orders.tot_units) AS total_orders,
    SUM(CASE WHEN orders.order_type = 'buy' then orders.tot_units Else 0 End) AS total_sold,
    SUM(CASE WHEN orders.order_type = 'return' then orders.tot_units Else 0 End) AS total_return
FROM product_dim AS products

INNER JOIN order_dim AS orders
    ON products.product_id = orders.product_id
    
GROUP BY products.brand

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q13.</b> How many distinct customers and delivery boys are there in each state?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    states.state,
    COUNT(DISTINCT customers.cust_id) AS Cutomers,
    COUNT(DISTINCT postman.delivery_person_id) AS Delivery_boys
FROM customer_dim AS customers 

INNER JOIN pincode_dim AS states 
    ON customers.primary_pincode = states.pincode
INNER JOIN delivery_person_dim AS postman
    ON postman.pincode = states.pincode

GROUP BY states.state
---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

```python
ans1 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    states.state,
    COUNT(DISTINCT postman.delivery_person_id) AS Delivery_boys
FROM delivery_person_dim AS postman

INNER JOIN pincode_dim AS states 
    ON postman.pincode = states.pincode 

GROUP BY states.state
""")

ans2 = ps.sqldf("""
SELECT
    states.state,
    COUNT(DISTINCT customers.cust_id) AS Cutomers
FROM customer_dim AS customers 

INNER JOIN pincode_dim AS states 
    ON customers.primary_pincode = states.pincode
    
GROUP BY states.state

---------------------WRITE QUERY ABOVE---------------------
""")

ans2.merge(ans1)
```

<b>Q14.</b> For every customer, print how many total units were ordered, how many units were ordered from their primary_pincode and how many were ordered not from the primary_pincode. Also calulate the percentage of total units which were ordered from primary_pincode(remember to multiply the numerator by 100.0). Sort by the percentage column in descending order.

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    customers.cust_id,
    customers.first_name,
    customers.last_name,
    SUM(orders.tot_units) AS total_units_ordered,
    SUM(CASE WHEN customers.primary_pincode = orders.delivery_pincode THEN orders.tot_units ELSE 0 END) AS From_primary_pincode,
    SUM(CASE WHEN customers.primary_pincode != orders.delivery_pincode THEN orders.tot_units ELSE 0 END) AS Non_primary_pincode,
    100.0 * SUM(CASE WHEN customers.primary_pincode = orders.delivery_pincode THEN orders.tot_units ELSE 0 END) / SUM(orders.tot_units) AS Percentage_same_city 
FROM customer_dim AS customers

INNER JOIN order_dim AS orders
    ON customers.cust_id = orders.cust_id

GROUP BY customers.cust_id

ORDER BY percentage_same_city DESC
---------------------WRITE QUERY ABOVE---------------------
""")

ans
```
